## Types of Conditional Statements

## 1.if Statement
Executes code only if the condition is true.

```python
marks=85
if marks>= 40:
    print("Pass")
```

## 2. if...else Statement

Executes one block if the condition is true and another if it is false.

```python
number=4
if number%2== 0:
    print("Even")
else:
    print("Odd")
```

## 3. if...elif...else

Used when checking multiple conditions.

```python
marks=72

if marks>=90:
    print("Grade A")
elif marks>=75:
    print("Grade B")
elif marks>=60:
    print("Grade C")
else:
    print("Grade D")
```

## 4. Nested if

A Nested if statement is a decision-making structure where one if statement is placed inside another if statement to evaluate multiple related conditions.

```python
age=20
citizen=True

if age>=18:
    if citizen:
        print("You are eligible to vote.")
```

## Some Experiment on conditional statement

```python
# Temperature
temperature=60
if temperature>30:
    print("It's really hot today")
elif temperature==30:
    print("Temperature is 30 degrees")
elif 20<temperature<30:
    print("Temperature is cool")
else:
    print("It's freezing")
```

```python
# Voting
age=19
if age>=18:
    print("You are eligible for vote")
else:
    print("You are not eligible for vote")
```

```python
# Grade
marks=50
if marks>=90:
    print("Grade A")
elif marks>=75:
    print("Grade B")
elif marks>=55:
    print("Grade C")
else:
    print("Fail")
```

```python
# Traffic light
color="red"
if color=="green":
    print("Go")
elif color=="yellow":
    print("Wait")
elif color=="red":
    print("Stop")
else:
    print("Invalid color")
```
