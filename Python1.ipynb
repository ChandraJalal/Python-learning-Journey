```python
python_class = 20.3
ml_class = 7.8
sql_class = 45.9
print("No of students in Python class:",python_class)
print("No of students in ML class:",ml_class)
print("No of students in SQL class:",sql_class)
print("Total number of Student:",python_class+ml_class+sql_class)
print("No of students in (python + sql):",python_class+sql_class)
print("No of students in(ML + sql):",ml_class+sql_class)
```
